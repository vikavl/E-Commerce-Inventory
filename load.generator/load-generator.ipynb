## Generate Orders Load

```python
import requests
import random
from concurrent.futures import ThreadPoolExecutor

BASE_URL = "http://<ALB-DNS>/api/v1"

def create_order():
    products = requests.get(f"{BASE_URL}/products").json()

    items = []
    for p in random.sample(products, k=random.randint(1, 3)):
        items.append({
            "productId": p["id"],
            "quantity": random.randint(1, 3)
        })

    payload = {"items": items}

    r = requests.post(f"{BASE_URL}/orders", json=payload)
    return r.status_code

def run_load(concurrency, requests_count):
    with ThreadPoolExecutor(max_workers=concurrency) as executor:
        futures = [executor.submit(create_order) for _ in range(requests_count)]
        for f in futures:
            f.result()

if __name__ == "__main__":
    run_load(concurrency=50, requests_count=500)
```
